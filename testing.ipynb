```python
import pandas as pd
import numpy as np

# Define the timeline
YEARS = 3
quarters_per_year = 4
total_quarters = YEARS * quarters_per_year
quarterly_dates = pd.date_range(start="2021-01-01", periods=total_quarters, freq='Q')

# Simulate batch data
np.random.seed(42)  # for reproducibility
batches_produced = np.random.normal(loc=110, scale=10, size=len(quarterly_dates)).astype(int)

# Different rejection rates before and after the quality control implementation
rejection_rate_pre = 0.15  # 15% rejection before
rejection_rate_post = 0.05  # 5% rejection after

rejection_rates = np.array([rejection_rate_pre]*quarters_per_year + [rejection_rate_post]*quarters_per_year +
                           [rejection_rate_pre]*quarters_per_year)
batches_rejected = (batches_produced * rejection_rates).astype(int)

# Costs and revenues
cost_per_batch = 500  # Cost per batch
cost_of_production = batches_produced * cost_per_batch
cost_of_rejected_batches = batches_rejected * cost_per_batch
net_production_cost = cost_of_production - cost_of_rejected_batches

# Revenue from batches
revenue_per_batch = 800
revenue = (batches_produced - batches_rejected) * revenue_per_batch
net_revenue = revenue - cost_of_rejected_batches

# Create DataFrame
quality_control_data = pd.DataFrame({
    "Quarter and Year": "Q"+ quarterly_dates.quarter.astype(str) + " " + quarterly_dates.year.astype(str),
    "Number of Batches Produced": batches_produced,
    "Number of Batches Rejected": batches_rejected,
    "Cost of Production": cost_of_production,
    "Cost of Rejected Batches": cost_of_rejected_batches,
    "Net Production Cost": net_production_cost,
    "Revenue": revenue,
    "Net Revenue": net_revenue
})
```
